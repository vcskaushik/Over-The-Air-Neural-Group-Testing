```python
def latex_print(df):
    for i in range(df.shape[0]):
        prt_str = ''
        curr_scheme = df["Scheme"][i]
        for col_j in df.keys():
            if (col_j == "Noise level")and (curr_scheme == "ITIT"):
                if df[col_j][i] =="No Noise":
                    prt_str += '\multirow{4}{1cm}{'+df[col_j][i]+'}'
                else:
                     prt_str += '\multirow{4}{1cm}{'+df[col_j][i][-1]+' dB}'
            elif col_j == "Scheme":
                prt_str += f' & {df[col_j][i]}'
#             elif (col_j == 'Group Tests') or (col_j == 'Total Tests') or (col_j == 'Total TMACs'):
#                 prt_str += f' & ${int(df[col_j][i])}$'
            elif (col_j == 'Group Tests') or (col_j == 'Total Tests'):
                prt_str += f' & ${df[col_j][i]:0.4g}$'
            
            elif col_j == 'Total Channel Uses':
                prt_str += f' & ${df[col_j][i]/(10**5) :0.2g} \cdot 10^5$'
            elif col_j != "Noise level":
                prt_str += f' & ${df[col_j][i]:0.2f}$'
            else:
                continue
        print(prt_str+ " \\\ ")
        if (curr_scheme== df["Scheme"][3]):
            print("\Xhline{1pt}")
        else:
            print("\cline{2-11}")
```
