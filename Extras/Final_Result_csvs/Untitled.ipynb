```python
def rm_txt_file(file_in,file_out,text_removed):
    with open(file_in,'r') as f1, open(file_out,'w') as f2:
        for line in f1:
            if text_removed in line :
                continue
            else:
                f2.write(line)
                
def compute_conf_mat(temp_list0,temp_list1):
    S0 = float(temp_list0[-1])
    S1 = float(temp_list1[-1])
    R0 = float(temp_list0[2])
    R1 = float(temp_list1[2])
    TP = R1*S1
    TN = R0*S0
    FP = (1-R0)*S0
    FN = (1-R1)*S1
    P = TP/(TP+FP)
    R = TP/(TP+FN)
    FPR = FP/(TN+FP)
    #ACC = (TP+TN)/(S0+S1)
    return P,R,FPR

                
def extract_metrics(file_in):
    train_acc_list = []
    val_acc_list = []
    val_p_list = []
    val_r_list = []
    val_fpr_list = []
    
    with open(file_in,'r') as f1:
        line_list = f1.read().splitlines()
        for i,line in enumerate(line_list):
            if "Epoch:" in line and "[242/243]" in line:
                temp_list = line.split(" ")
                item = temp_list[-1].split(")")
                try:
                    item_float = float(item[0])
                except:
                    print("200 epochs not completed",item[0])
                    break
                train_acc_list += [item_float]
                
            if "VAL * Acc@1" in line:
                temp_list = line.split(" ")
                try:
                    item_float = float(temp_list[3])
                except:
                    print("90 epochs not completed",temp_list[3])
                    break
                #print(item[0])
                #print(temp_list[3])
                val_acc_list += [item_float]
                
            if "classification_report" in line:
                temp_list0 = (" ".join(line_list[i+3].split())).split(" ")
                temp_list1 = (" ".join(line_list[i+4].split())).split(" ")
                P,R,FPR = compute_conf_mat(temp_list0,temp_list1)
                val_p_list += [P]
                val_r_list += [R]
                val_fpr_list += [FPR]
            
    return train_acc_list,val_acc_list,val_p_list,val_r_list,val_fpr_list
```

```python
def confmat2metrics(conf_mat,Model_name):
    tn, fp, fn, tp = conf_mat.ravel() 
    recall = tp/(tp+fn)
    fpr = fp/(tn+fp)
    acc = (tp+tn)/np.sum(conf_mat)
    print("************************************")
    print(Model_name)
    print("************************************")
    print("Accuracy : ",acc)
    print("Recall :",recall)
    print("False Positive Rate",fpr)
    print("************************************")
    
```
